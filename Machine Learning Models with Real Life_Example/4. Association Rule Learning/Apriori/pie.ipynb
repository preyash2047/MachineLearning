```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```python
dataset = pd.read_csv('DV_sample.csv')
```

```python
dataset.head()
```

# Pie

```python
del dataset["Date"]
data = np.array(dataset.sum())

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data, autopct = lambda pct: func(pct, data), labels = dataset.columns)
  
plt.setp(autotexts, size = 8)
plt.show()
```
